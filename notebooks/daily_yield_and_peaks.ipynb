```python
import pandas as pd

dfs = pd.read_csv(f'.././datasets/tsx/DH_div_history_T.csv', keep_default_na=False)

dfs2 = pd.read_csv(f'.././datasets/price_history/yahoo_price_history_T.TO.csv')

```

```python
dfs.rename(columns={'Ex-Dividend Date':'Date'}, inplace=True)
result = pd.merge(dfs, dfs2, on=['Date'], how='outer')
result.sort_values(by='Date', inplace=True)



result.fillna(method='ffill', inplace=True)
#result.fillna(method='bfill', inplace=True)

print(dfs.columns)
print(dfs2.columns)
result.drop(columns=['Close', 'High', 'Low', 'Dividends', 'symbol', 'market', 'date_epoch'], inplace=True)
print(result.columns)
print(result.loc[5])
print(result.loc[21])
print(result)
```

```python
# need frequency
# first go does not account for change in frequency for single stocks

dfs3 = pd.read_csv(f'.././datasets/weekly_divhistory_reports/CAN/div_history_report-CAN-2020-08-14.csv', keep_default_na=False)

print(dfs3.head())

# create frequency lookup table
dfs3.drop(columns=["Name", "Price", "Yld", "Ex-Div", "PayRto", "PE", "PB", "Beta", "Mkt Cap", "WK%", "MO%", "2MO%", "3MO%", "6MO%", "1YR%", "report_date_epoch", "ex_div_epoch"], inplace=True)

print(dfs3)

# pull frequency from table

freq = dfs3.loc[dfs3['Symbol']=='T']
print(freq)
```

```python
# got freq now calculate yield for all prices
factor = 0

if (freq.iloc[0]['div_freq'] == 'Q'):
    factor = 4
print(factor)
```

```python
# iterate through results and add div yield for each row
# daily_yield = factor*result[Cash Amount]/result[Close]

#for index, row in result.iterrows():
    #print(index)
    #print(row)
    #daily_yield = factor*result.iloc[index]['Cash Amount']/result.iloc[index]['Close']
result['Daily Yield'] = factor*result['Cash Amount']/result['Open']
print(result[3500:4000]['Open'])
```

```python
print(result.columns)
result = result[result['Daily Yield'].notna()]
#print(subset)
```

```python
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objects as go
import cufflinks as cf
init_notebook_mode(connected=True)

trace = go.Scatter(x = result['Date'], y=result['Daily Yield'], text = result['Open'], mode='markers')
data=[trace]

layout = go.Layout(title="Daily Yield over time", xaxis=dict(title='Date'), yaxis=dict(title='Yield'), hovermode='closest')

figure = go.Figure(data=data, layout=layout)
iplot(figure)
```

```python
from scipy.signal import find_peaks
import pandas as pd
import plotly.graph_objects as go
import numpy as np
import matplotlib.pyplot as plt

dfs5 = pd.read_csv('../datasets/yield_history/yield_history_tsx_REI.UN.csv')
#print(dfs5)

value_array = dfs5['Daily Yield']
date_array = dfs5['Date']
#print(value_array)
peaks = find_peaks(value_array, prominence=0.004)[0]
valleys = find_peaks(-value_array, prominence=0.004)[0]
#print(type(peaks))
#print(peaks[1])
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=date_array,
    y=value_array,
    mode='lines+markers',
    name='Original Plot'
))

fig.add_trace(go.Scatter(
    x=peaks,
    y=[value_array[j] for j in peaks],
    mode='markers',
    marker=dict(
        size=8,
        color='red',
        symbol='cross'
    ),
    name='Detected Peaks'
))

fig.add_trace(go.Scatter(
    x=valleys,
    y=[value_array[j] for j in valleys],
    mode='markers',
    marker=dict(
        size=8,
        color='green',
        symbol='cross'
    ),
    name='Detected Valleys'
))

fig.update_xaxes(date_array)

fig.show()

total =0 
for temp in peaks:
    total += value_array[temp] 
    
peak_avg = total/len(peaks)
print(peak_avg)

total =0 
for temp in valleys:
    total += value_array[temp] 
    
valley_avg = total/len(valleys)
print(valley_avg)
#plt.subplot(2, 2, 1)
#plt.plot(peaks, value_array[peaks], "xr"); plt.plot(x); plt.legend(['prominence'])
#plt.subplot(2, 2, 2)
#plt.plot(peaks2, x[peaks2], "ob"); plt.plot(x); plt.legend(['prominence'])
#plt.subplot(2, 2, 3)
##plt.plot(peaks3, x[peaks3], "vg"); plt.plot(x); plt.legend(['width'])
#plt.subplot(2, 2, 4)
#plt.plot(peaks4, x[peaks4], "xk"); plt.plot(x); plt.legend(['threshold'])
#plt.show()
```
